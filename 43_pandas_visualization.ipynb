```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9], 'D': [10, 11, 12]}
df = pd.DataFrame(data)
df
```

```python
df.plot()
plt.show()
```

```python
df['A'].plot(kind='line')
plt.show()
```

```python
df['B'].plot(kind='bar')
plt.show()
```

```python
df['B'].plot(kind='hist')
plt.show()
```

```python
df.plot(x='A', y='B', kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['A'].plot(kind='line', linestyle='--', marker='*', color='red')
plt.show()
```

```python
df['B'].plot(kind='bar')
plt.title('Value of B')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
flg, axes = plt.subplots(nrows=2, ncols=2)
df['A'].plot(ax=axes[0, 0])
df['A'].plot(kind='bar', ax=axes[0, 1])
df['B'].plot(ax=axes[1, 0])
df['B'].plot(kind='bar', ax=axes[1, 1])

plt.show()
```

```python
df['A'].plot(kind='box')
plt.show()
```

```python
df.plot(kind='box')
plt.show()
```

```python
import geopandas as gpd
```

```python
data = {
    'longitude': [-74.006, -118.2437, 2.3522],
    'latitude': [40.7128, 34.0522, 48.8566]
}

geodata = pd.DataFrame(data)
geodata
```

```python
gdf = gpd.GeoDataFrame(geodata, geometry=gpd.points_from_xy(geodata['longitude'], geodata['latitude']))
```

```python
url = 'https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip'
worlds = gpd.read_file(url)

worlds.plot()
gdf.plot(ax=plt.gca(), marker='o', color='red', markersize=50)
plt.show()
```
