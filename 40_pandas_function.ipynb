```python
from sklearn.datasets import make_regression
import pandas as pd
```

```python
x, y = make_regression(n_samples=100, n_features=1, noise=5, random_state=42)
x
```

```python
len(y)
```

```python
data = pd.DataFrame({'X': x.flatten(), "Y": y})
data
```

```python
data.info()
```

```python
window_size = 2

rolling_window = data['Y'].rolling(window=window_size)
rolling_window
```

```python
rolling_window.mean()
```

```python
data['Rolling_mean_y'] = rolling_window.mean()
data
```

```python
data['Rolling_sum_y'] = rolling_window.sum()
data
```

```python
data['Rolling_min_y'] = rolling_window.min()
```

```python
data
```

```python
window_size = 3

grouped_data = data.groupby('Group')['Y'].rolling(window=window_size).sum()
grouped_data
```

```python
def custom_mean(window):
    return window.mean(skipna=True)

window_size = 5
custom_window = data['Y'].rolling(window=window_size)

data['Custom Rolling mean'] = custom_window.apply(custom_mean)
data
```
